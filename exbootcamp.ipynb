```python
import pandas as pd
import numpy as np


# Caminho do arquivo
caminho_arquivo = '/content/sample_data/1KBtJf20QHGyrNOTEGeq_saude_do_sono_estilo_vida (1).csv'

# Leitura do arquivo CSV
df = pd.read_csv(caminho_arquivo)



# Renomeando colunas
df.rename(columns={'ID': 'Identificador', 'Pressão sanguíneaaaa': 'Pressão sanguínea', 'Categoria BMI': 'Categoria IMC', 'Ocupação': 'Profissão'}, inplace=True)

# 2. Média, moda e mediana de horas de sono para cada profissão
mean_sleep = df.groupby('Profissão')['Duração do sono'].mean()
mode_sleep = df.groupby('Profissão')['Duração do sono'].apply(lambda x: x.mode().iloc[0])
median_sleep = df.groupby('Profissão')['Duração do sono'].median()

# Exibindo os resultados
print("\nMédia de horas de sono para cada profissão:")
print(mean_sleep)
print("\nModa de horas de sono para cada profissão:")
print(mode_sleep)
print("\nMediana de horas de sono para cada profissão:")
print(median_sleep)

# 3. Porcentagem de obesos entre os engenheiros de software
engineer_obese_percentage = df[df['Profissão'] == 'Eng. de Software']['Categoria IMC'].value_counts(normalize=True).get('Obesidade', 0) * 100
print(f"A porcentagem de obesos entre os engenheiros de software é: {engineer_obese_percentage:.2f}%")

# 4. Advogados ou representantes de vendas dormem menos?
profissoes_interesse = ['Advogado(a)', 'Representante de Vendas']
mean_sleep_lawyers_sales = df[df['Profissão'].isin(profissoes_interesse)]['Duração do sono'].mean()
print(f"\nMédia de horas de sono para advogados e representantes de vendas: {mean_sleep_lawyers_sales:.2f}")

# 5. Quem tem menos horas de sono, enfermeiros ou médicos?
profissoes_interesse_2 = ['Enfermeiro(a)', 'Médico(a)']
mean_sleep_nurses_doctors = df[df['Profissão'].isin(profissoes_interesse_2)]['Duração do sono'].mean()
print(f"\nMédia de horas de sono para enfermeiros e médicos: {mean_sleep_nurses_doctors:.2f}")

# 6. Subconjunto de colunas
subset_cols = ['Identificador', 'Gênero', 'Idade', 'Pressão sanguínea', 'Frequência cardíaca']
df_subset = df[subset_cols]
print("\nSubconjunto com as colunas selecionadas:")
print(df_subset)

# 7. Qual a profissão menos frequente no conjunto.
least_frequent_profession = df['Profissão'].value_counts().idxmin()
print(f"A profissão menos frequente é: {least_frequent_profession}")

# 8. Quem tem maior pressão sanguínea média, homens ou mulheres? (Considerando a média)
pressure_extract = df['Pressão sanguínea'].str.extract(r'(\d+)/(\d+)')
df[['Pressão Mínima', 'Pressão Máxima']] = pressure_extract.astype(float)
mean_pressure_gender = df.groupby('Gênero')[['Pressão Mínima', 'Pressão Máxima']].mean()
print("\nMédia de pressão sanguínea para homens e mulheres:")
print(mean_pressure_gender)

# 9. Predominância de dormir 8 horas por dia (usando Moda)
mode_sleep = df['Duração do sono'].mode()
print(f"\nA moda de horas de sono é: {mode_sleep.iloc[0]}")

# 10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)
mean_steps_above_70 = df[df['Frequência cardíaca'] > 70]['Passos diários'].mean()
mean_steps_70_or_below = df[df['Frequência cardíaca'] <= 70]['Passos diários'].mean()
print(f"\nMédia de passos para frequência cardíaca acima de 70: {mean_steps_above_70:.2f}")
print(f"Média de passos para frequência cardíaca 70 ou menos: {mean_steps_70_or_below:.2f}")
```
